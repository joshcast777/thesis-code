# Passing the non-repeated data from `PRM_FS_TASK2_2_V1` to `PRM_ZS_TASK2_2_V1`

```python
# Import libraries

import pandas as pd
```

```python
# Function to count records

def show_record_count(df):
	"""Display the total number of records in a DataFrame.

	Args:
		df (pandas.DataFrame): The DataFrame for which the number of records will be displayed.

	Returns:
		str: A string indicating the total number of records in the DataFrame.
	"""

	return f"Total records: {len(df)}"
```

```python
# Constants

# Paths
DATA_PATH = "data/"
"""str: Path for the data."""

# File names
FEW_SHOT_FILE = "PRM_FS_TASK2_2_V1"
"""str: Name of the Few-Shot file."""

ZERO_SHOT_FILE = "PRM_ZS_TASK2_2_V1"
"""str: Name of the Zero-Shot file."""

FS_ZS_JOINED_FILE = "FS_ZS_JOINED"
"""str: Name of the Zero-Shot and Few-Shot file joined."""

# Fields
SNT_ID_FIELD = "snt_id"
"""str: Name of the snt_id field."""

TERM_FIELD = "term"
"""str: Name of the term field."""
```

```python
# Read Excel files

fs = pd.read_excel(f"{DATA_PATH}{FEW_SHOT_FILE}.xlsx", sheet_name="Sheet1")
zs = pd.read_excel(f"{DATA_PATH}{ZERO_SHOT_FILE}.xlsx", sheet_name="Sheet1")
```

```python
print(show_record_count(fs[SNT_ID_FIELD]))

fs.head()
```

```python
print(show_record_count(zs[SNT_ID_FIELD]))

zs.head()
```

```python
# Format data

fs[TERM_FIELD] = fs[TERM_FIELD].str.capitalize()
zs[TERM_FIELD] = zs[TERM_FIELD].str.capitalize()
```

```python
print(show_record_count(fs[SNT_ID_FIELD]))

fs.head()
```

```python
print(show_record_count(zs[SNT_ID_FIELD]))

zs.head()
```

```python
# Get FS data not repeated in ZS

result = fs[~fs.set_index([SNT_ID_FIELD, TERM_FIELD]).index.isin(zs.set_index([SNT_ID_FIELD, TERM_FIELD]).index)]

print(show_record_count(result[SNT_ID_FIELD]))

result.head()
```

```python
# Concat results data to ZS

fs_zs_joined = pd.concat([zs, result])

print(show_record_count(fs_zs_joined[SNT_ID_FIELD]))

fs_zs_joined.head()
```

```python
# Exporting as an Excel file

fs_zs_joined.to_excel(f"{DATA_PATH}{FS_ZS_JOINED_FILE}.xlsx", sheet_name=FS_ZS_JOINED_FILE, index=False)
```

The `FS_ZS_JOINED` file is the union of non-repeated data from `PRM_FS_TASK2_2_V1` to `PRM_ZS_TASK2_2_V1`
